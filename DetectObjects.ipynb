Za detekciju je izabran model dobiven treniranjem YOLOv8n modela, spremljen u folder train20 s najboljim te≈æinama.

```python
def Detect(p):
    model = YOLO('C:/Users/Ivona/Desktop/Codeasy/runs/detect/train20/weights/best.pt')
    results = model(image_path, conf=0.2, imgsz=640, show=True, show_labels=True, show_conf=True, show_boxes=True)
    for r in results:
        im_array = r.plot()         
        plt.imshow(im_array[..., ::-1])
        plt.axis('off')
        plt.tight_layout(pad=0)
        plt.show()
        
```
